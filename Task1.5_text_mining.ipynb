### 3. Word tokenization

```python
from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(mydata)
print(tokenized_word)
```

### 4. Stop words

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords from the tokenized_word list

filtered_words = []
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
print("Before:", len(tokenized_word))
print("After:", len(filtered_words))
```

```python
# Substitute all punctuation marks with a space 

cleaned_words = []
for word in filtered_words:
    cleaned_word = re.sub("[^a-zA-Z]", "", word)  # Remove non-letter characters
    if cleaned_word:  # Only add non-empty strings
        cleaned_words.append(cleaned_word)  
```

```python
print("Before:", len(tokenized_word))
print("After:", len(filtered_words))
print("After:",len (cleaned_words))
```

```python
# New word tokenisation

text = " ".join(cleaned_words) 
tokenized_word_new = word_tokenize(text)
print(tokenized_word_new)
```

```python
# Create a new FreqDist

freq_words_filter_new = FreqDist(tokenized_word_new)
```

```python
freq_words_filter_new=freq_words_filter_new.most_common(10)
```

```python
# convert to a dataframe 
df_words_new = pd.DataFrame(freq_words_filter_new, columns=['word', 'frequency'])
```

```python
## creating a plot 
plt.figure(figsize=(10, 6))
sns.barplot(x='word', y='frequency', data=df_words_new, palette='Blues_d')
plt.title('Top 10 Most Common Words')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join(output_folder, 'Top 10_common words.png'), dpi=300)
plt.show()
```

### After removing the stop words and punctuations, there is a huge difference in the plot. Earlier, most common words were punctuation marks and made no meaning. Now except 'The'and 's', most other words make slight sense in the key events of 20th century.

### 6. Top 15 POS labels each for noun, adjective and verb

```python
df_text.head()
```

```python
# Define tags for each group
noun_tags = ['NN', 'NNS', 'NNP', 'NNPS']
verb_tags = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
adj_tags = ['JJ', 'JJR', 'JJS']
```

```python
def top_words(df, tag_list, top_n=15):
    filtered_df = df[df['Word type'].isin(tag_list)]
    return (
        filtered_df.groupby('Words')
        .size()
        .reset_index(name='Occurrences')
        .sort_values(by='Occurrences', ascending=False)
        .head(top_n)
    )
```

```python
top_nouns = top_words(df_text, noun_tags)
```

```python
top_verbs = top_words(df_text, verb_tags)
```

```python
top_adjs = top_words(df_text, adj_tags)
```

```python
# Plotting function
def plot_top_words(df, title):
    plt.figure(figsize=(10, 6))
    plt.barh(df['Words'][::-1], df['Occurrences'][::-1], color='skyblue')  # reverse for descending order
    plt.title(title)
    plt.xlabel('Occurrences')
    plt.ylabel('Words')
    plt.tight_layout()
```

```python
# Plot each category
# Top 15 nouns
plot_top_words(top_nouns, 'Top 15 Nouns')
plt.tight_layout()
plt.savefig(os.path.join(output_folder, 'Top 15 nouns.png'), dpi=300)
plt.show()
```

```python
plot_top_words(top_verbs, 'Top 15 Verbs')  ## verbs
plt.savefig(os.path.join(output_folder, 'Top 15 verbs.png'), dpi=300)
```

```python
plot_top_words(top_adjs, 'Top 15 Adjectives')  ## adjectives
plt.savefig(os.path.join(output_folder, 'Top 15 adjectives.png'), dpi=300)
```

### The highest occurences are of nouns as compared to verbs and adjectives. The adjectives contain country names.

```python
## Import countries file
df_countries=pd.read_csv(r'C:\Users\akul_\OneDrive\Documents\Tarishi\CF Data analysis bootcamp\Data visualizations with Python (Specialization)\Achievement 1\20th-century\countries_list_20th_century_1.5.csv')
```

```python
df_countries.head()
```

```python
country_list = df_countries['country_name'].dropna().tolist()
country_list = [c.strip() for c in df_countries['country_name'].dropna().tolist()]   ##fix extra spaces in country csv file
```

```python
## Importing file 
with open('key_events_Wiki.txt', 'r', errors='ignore') as file:
    mydata = file.read().replace('\n', '')
```

```python
# Load the text and convert to plain string if it's a TextBlob
blob = TextBlob(mydata)
text = str(blob)  # This avoids the TypeError
```

```python
for country_name in country_list:
    pattern = r'\b' + re.escape(country_name) + r'\b'
    matches = re.findall(pattern, text, flags=re.IGNORECASE)
    print(f"{country_name}: {len(matches)}")  # Debug print
    if matches:
        country_counts[country_name] = len(matches)
```

```python
# Convert to DataFrame
df = pd.DataFrame(list(country_counts.items()), columns=['Country', 'Mentions'])
df = df.sort_values(by='Mentions', ascending=False).reset_index(drop=True)
print(df)
```

```python
visualizations= 'output_figures'
file_name = 'mentions_by_country.png' 
```

```python
# Plot
plt.figure(figsize=(12, 6))
plt.bar(df['Country'], df['Mentions'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Frequency of Country Mentions in Text')
plt.xlabel('Country')
plt.ylabel('Number of Mentions')
plt.tight_layout()
plt.savefig(os.path.join(output_folder, 'country mentions.png'), dpi=300)
plt.show()
```

### The plot shows that countries-Germany, Japan, United States and France are mentioned the most in the text. This looks good as these countries were involved in the major events of 20th century, which was the world war. While other countries are mentioned once or twice from the list.
